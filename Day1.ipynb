```python
df["Cabin"].isnull().sum()
df["Cabin"]=df["Cabin"].fillna("Unknown")
df["Deck"] = df["Cabin"].str[0]
df
```

```python
df["Sex"]= df["Sex"].map({"male" : 1, "female":0})
df
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
plt.hist(df['Fare'], bins=30, alpha=0.7, color='g', edgecolor='black')
plt.title("Original Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Frequency")
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df['Fare_normalized'] = scaler.fit_transform(df[['Fare']])

# Step 3: Visualize the normalized Fare column
plt.subplot(1, 2, 2)
plt.hist(df['Fare_normalized'], bins=30, alpha=0.7, color='b', edgecolor='black')
plt.title("Normalized Fare Distribution")
plt.xlabel("Normalized Fare")
plt.ylabel("Frequency")
```
